# Import Vars

```python
import supervised_ml
# order of scores in each list:
# (accuracy, f1, precision, recall)
rfc_computed = supervised_ml.rfc_computed
erf_computed = supervised_ml.erf_computed
lr_computed = supervised_ml.lr_computed
svm_computed = supervised_ml.svm_computed
knn_computed = supervised_ml.knn_computed
```

```python
mega_ls = [rfc_computed, erf_computed, lr_computed, svm_computed, knn_computed]
```

```python
# dependencies
import matplotlib.pyplot as plt
import numpy as np
```

# Graphing

```python
# setting up vars
model_names = ['Random Forest', 'E. Random Forest', 'Log Regression', 'SVM']
color_ls = ['green', 'orange', 'blue', 'teal']
scores_ls = ['Accuracy', 'F1', 'Precision', 'Recall']

stats_acc = [round(mega_ls[i][0],3) for i in range(0,len(mega_ls)-1)]
stats_f1 = [round(mega_ls[i][1],3) for i in range(0,len(mega_ls)-1)]
stats_prec = [round(mega_ls[i][2],3) for i in range(0,len(mega_ls)-1)]
stats_rec = [round(mega_ls[i][3],3) for i in range(0,len(mega_ls)-1)]
big_stats = [stats_acc, stats_f1, stats_prec, stats_rec]
```

```python
#mega setup
fig, ((gr0, gr1), (gr2, gr3)) = plt.subplots(2, 2, figsize = (10,13))
fig.suptitle("Score Comparisons of All Models", fontweight = 'bold')
fig.supxlabel("Model Type", fontweight = 'bold')
fig.supylabel("Model Score", fontweight = 'bold')

#plotting graph
for i in range(len(model_names)):
    #plot graph
    eval(f"gr{i}").set_xticklabels(model_names, rotation=15)
    eval(f"gr{i}").bar(model_names, big_stats[i], color=color_ls[i])

    #set plot features
    eval(f"gr{i}").grid(alpha = 0.25)
    eval(f"gr{i}").title.set_text(f'{scores_ls[i]} Scores')

    #annotate values
    for index, v in enumerate(big_stats[i]):
        eval(f"gr{i}").text(index-.15, v+.005, str(v))

plt.savefig('Resources/Export/model_score_compare.png')
plt.tight_layout()
```
