```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

# Data Preprocessing

```python
#import data
import pandas as pd
df = pd.read_csv('drive/My Drive/bootcamp_colab/Resources/secondary_data.csv', sep=';')
df.head()
```

```python
#check num of null values
df.isnull().sum()
```

```python
#drop features with more than 75% null values
df_smaller = df.drop(columns={'stem-root', 'stem-surface','veil-type', 'veil-color', 'spore-print-color', 'gill-spacing'})
df_smaller.head()

#dropped columns that had large number of nan
#nan could be a useful classifier, will consider re-running with the nans left in
```

```python
#dropping rows with nans
df_smaller.dropna(inplace=True)
```

```python
#one-hot encoding
df_dummies = pd.get_dummies(df_smaller)
df_dummies
```

```python
#separate X and y data
X = df_dummies.drop(columns = {'class_e','class_p'})
y = df_dummies['class_e']
```

```python
# splitting dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# scaling the data
# using standard scaler because assuming the mushroom data is normal
X_scaler = StandardScaler()
X_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Modeling

### Random Forest

```python
# random forest function for modularization
def rfc(X_train, X_test, y_train, y_test):
  clf = RandomForestClassifier(n_estimators=50, random_state=42)
  clf.fit(X_train, y_train)
  score = clf.score(X_test, y_test)
  y_pred_Log = clf.predict(X_test)
  print(classification_report(y_test, y_pred_Log))
  # clf.predict(X_test)
  return score
```

```python
rfc(X_train_scaled, X_test_scaled, y_train, y_test)
```

### Logistic Regression

```python
# logistic regression function for modularization
def lr(X_train, X_test, y_train, y_test):
  model = LogisticRegression()
  model.fit(X_train, y_train)
  score = model.score(X_test, y_test)
  y_pred_Log = model.predict(X_test)
  print(classification_report(y_test, y_pred_Log))
  return score
```

```python
lr(X_train_scaled, X_test_scaled, y_train, y_test)
```
